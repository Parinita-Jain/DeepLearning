```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils #for onehotencoding

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
df.head()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y = np_utils.to_categorical(y)
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
ann = Sequential()
ann.add(Dense(units=8, activation="relu"))
ann.add(Dense(units=4, activation="relu"))
ann.add(Dense(units=3, activation="softmax")) #output
ann.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
ann.fit(xtrain, ytrain, epochs=200)
```

```python
pred = ann.predict(xtest)
length = len(pred)

ylabel = np.argmax(ytest, axis=1) #actual
predlabel = np.argmax(pred, axis=1) #predicted

ac = np.sum(ylabel==predlabel)/length * 100

print(f"Accuracy : {ac}")
```

```python
def mymodel():
  ann = Sequential()
  ann.add(Dense(units=8, activation="relu"))
  ann.add(Dense(units=4, activation="relu"))
  ann.add(Dense(units=3, activation="softmax")) #output
  ann.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
  return ann
```

```python
estimator = KerasClassifier(build_fn= mymodel, epochs=200, batch_size=5, verbose=0 )
cvs = cross_val_score(estimator, x,y, cv=10)
print(f"Avg Accuracy -: {cvs.mean()*100}\nSTD : {cvs.std()}")

```
