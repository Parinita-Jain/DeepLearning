```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import mean_squared_error
```

```python
df =pd.read_csv("fake_reg.csv")
df.head()
```

```python
df.info()
```

```python
x = df.iloc[:,1:]
y = df.iloc[:, 0]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
#step1 -: initialise the ann model
ann = Sequential()

#step2 -: add layers into the model
ann.add(Dense(units=4, activation="relu"))

ann.add(Dense(units=1))  #in regression , no activation func for o/p 

#step3 -: establish the connection between the layers
ann.compile(optimizer="adam", loss="mse") # aslo there is no metrics parameter

#step4 -: train the model
ann.fit(xtrain, ytrain, batch_size=20, epochs=500)

#step5 -: make prediction
ypred = ann.predict(xtest)
```

```python
print(f"MSE -: {mean_squared_error(ytest, ypred)}")
print(f"RMSE -: {np.sqrt(mean_squared_error(ytest, ypred))}") 
```

```python
ann.evaluate(xtrain, ytrain)
```

```python
ann.evaluate(xtest, ytest)
```
